```python
import pandas as pd 

clean_df = pd.read_csv('clean_tabla.csv', index_col=0)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import umap
import umap.plot
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline

df = clean_df.copy()
df.head(5)
```

```python
not_nan = (df.mate_nc_na == 0) & (df.lengua_nc_na == 0) & (df.naturales_nc_na == 0) & (df.sociales_nc_na == 0)
train = df[not_nan]

#test = df[~not_nan]
```

```python
to_keep = ['naturales_Bajo_Basico','naturales_Basico', 'naturales_satisfactorio', 'naturales_avanzado', 'sociales_Bajo_Basico', 'sociales_Basico'
  , 'sociales_satisfactorio', 'sociales_avanzado', 'lengua_Bajo_Basico', 'lengua_Basico', 'lengua_satisfactorio', 'lengua_avanzado',
  'Mate_Bajo_Basico', 'Mate_Basico', 'Mate_satisfactorio', 'Mate_avanzado'
]
train = train[to_keep]
```

```python
train.describe()
```

```python
data = train.sample(2250).copy()
```

```python
from sklearn.preprocessing import MinMaxScaler
min_max_scaler = MinMaxScaler()
scaled_values = min_max_scaler.fit_transform(data.values)
data = pd.DataFrame(data=scaled_values, index=data.index, columns=data.columns)
```

```python
data.describe()
```

```python
def label(row):
  if row['Mate_Bajo_Basico'] > 0:
    return 0
  if row['Mate_Basico'] > 0:
    return 1
  if row['Mate_satisfactorio'] > 0:
    return 2
  if row['Mate_avanzado'] > 0: 
    return 3
  else:
    return -1
```

```python
def plot_umap(df, n_neighbors=15, min_dist=0.1,spread=1.0, metric='euclidean'):
  mapper = umap.UMAP(n_neighbors=n_neighbors, min_dist=min_dist,metric=metric)
  mapper.fit(df.copy())
  labels = df.copy().apply(lambda x: label(x), axis=1)
  umap.plot.points(mapper, labels=labels, background='black')
```

```python
%%time
size = [1.1, 1.2, 1.3, 1.4, 1.5]
for i in range(len(size)):
  plot_umap(df=data,n_neighbors=200,min_dist=0.6, spread=size[i], metric='seuclidean')
```
