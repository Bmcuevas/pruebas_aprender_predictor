```
import pandas as pd 

alumno_df = pd.read_csv('/content/drive/MyDrive/Hackaton/clean-data/Alumno.csv', index_col=0)
tabla_df = pd.read_csv('/content/drive/MyDrive/Hackaton/clean-data/Tabla.csv', index_col=0)
depto_df = pd.read_csv('/content/drive/MyDrive/Hackaton/clean-data/Depto.csv', index_col=0)
escuela_df = pd.read_csv('/content/drive/MyDrive/Hackaton/clean-data/Escuela.csv', index_col=0)
provincia_df = pd.read_csv('/content/drive/MyDrive/Hackaton/clean-data/Provincia.csv', index_col=0)
```

```
df = tabla_df.copy()
```

```
import numpy as np


df = df.replace('nc', np.nan)
df.isnull().sum()
```

```
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

**nan = -1,
Yes aid = 1,
No aid = 2**

```
%%capture
df.subvencion[df.subvencion.isnull()] = np.nan
df.subvencion[df.subvencion == 'Sin información'] = np.nan
df.subvencion[df.subvencion == 'Subvención Total'] = 1
df.subvencion[df.subvencion == 'Más de 75% y hasta 99%'] = 1
df.subvencion[df.subvencion == 'Más de 50% y hasta 75%'] = 1
df.subvencion[df.subvencion == 'Más de 25% y hasta 50%'] = 1
df.subvencion[df.subvencion == 'Hasta 25%'] = 1
df.subvencion[df.subvencion == 'Parcial'] = 1
df.subvencion[df.subvencion == 'No'] = 2
df.subvencion[df.subvencion == 'No corresponde'] = 2
```

```
%%capture
df.gestion[df.gestion == 'estatal'] = 0
df.gestion[df.gestion == 'privada'] = 1
df.gestion[df.gestion == 'social/cooperativa'] = 2
```

```
df[df.gestion == 2].count()
```

**Remove all records that gestion == social/cooperativa, we only lose 750**

```
df = df[df.gestion != 2]
```

```
%%capture
df.indice_socioeconomico[df.indice_socioeconomico.isnull()] = np.nan
df.indice_socioeconomico[df.indice_socioeconomico == 'bajo'] = 0
df.indice_socioeconomico[df.indice_socioeconomico == 'medio'] = 1
df.indice_socioeconomico[df.indice_socioeconomico == 'alto'] = 2
```

```
%%capture
df.indice_socioeconomico_medio_cat[df.indice_socioeconomico_medio_cat.isnull()] = np.nan
df.indice_socioeconomico_medio_cat[df.indice_socioeconomico_medio_cat == 'tercil_bajo'] = 0
df.indice_socioeconomico_medio_cat[df.indice_socioeconomico_medio_cat == 'tercil_medio'] = 1
df.indice_socioeconomico_medio_cat[df.indice_socioeconomico_medio_cat == 'tercil_alto'] = 2
```

**Femail = 0,  Male = 1**

```
%%capture
df.sexo[df.sexo == 'f'] = 0
df.sexo[df.sexo == 'm'] = 1
```

**We can get rid of nivel_desemp categorical features and ponderador nominal features since their value is implied by their hot encoded column**

```
%%capture
df.drop(columns=['ponderador_lengua','ponderador_matematica','ponderador_ciencias_sociales','ponderador_ciencias_naturales'], axis=1, inplace=True)
```

**Format all tecnology columns: Si tiene = 1, No tiene = 0**

```
si_replace = {
    'tiene_notebook': 'si',
    'tiene_pc': 'si',
    'tiene_tablet': 'si',
    'tiene_celular': 'si',
    'tiene_smartphone': 'si',
    'tiene_consola': 'si',
    'tiene_smarttv': 'si',
    'tiene_cable': 'si',
    'tiene_internet': 't'
}

df.replace(to_replace=si_replace, value=1, inplace=True)

no_replace = {
    'tiene_notebook': 'no',
    'tiene_pc': 'no',
    'tiene_tablet': 'no',
    'tiene_celular': 'no',
    'tiene_smartphone': 'no',
    'tiene_consola': 'no',
    'tiene_smarttv': 'no',
    'tiene_cable': 'no',
    'tiene_internet': 'f'
}


df.replace(to_replace=no_replace, value=0, inplace=True)
```

**Format repeticion_primaria y repeticion_secundaria columns: No = 0, Una_vez = 1, dos_veces = 2, tres_veces_o_mas = 3**

```
replace = {
    'repeticion_primaria':{
        'no': 0, 'una_vez': 1, 'dos_veces': 2, 'tres_veces_o_mas': 3
    },
    'repeticion_secundaria':{
        'no': 0, 'una_vez': 1, 'dos_veces': 2, 'tres_veces_o_mas': 3
    }
}

df.replace(to_replace=replace, value=None, inplace=True)
```

**Format ambito column: urbano = 0, rural = 1**

```
ambito_replace = {
    'ambito': {
        'urbano': 0,
        'rural': 1
    }
}

df.replace(to_replace=ambito_replace, value=None, inplace=True)
```

**Format dependencia column: provincial = 0, nacional = 1, municipal = 2**

```
dep_replace = {
    'dependencia': {
        'provincial': 0,
        'nacional': 1,
        'municipal': 2,
        'sin información': np.nan
      }
}

df.replace(to_replace=dep_replace, value=None, inplace=True)
```

**Format icse_cat column: bajo_y_medio = 0, alto = 1, critico = 2**

```
icse_replace = {
    'icse_cat': {
        'bajo_y_medio': 0,
        'alto': 1,
        'critico': 2
    }
}

df.replace(to_replace=icse_replace, value=None, inplace=True)
```

**Format tecnica column: false = 0, true = 1**

```
tecnica_replace = {
    'tecnica': {
        'f': 0,
        't': 1
    }
}

df.replace(to_replace=tecnica_replace, value=None, inplace=True)
```

**Mate_nc_na is a repeated column, so we drop it**

```
df.drop(columns=['Mate_nc_na'], axis=1, inplace=True)
```

**Replace departamento and provincia columns with their IDs**

```
depto_copy = depto_df.copy()
depto_copy.rename(columns={'id':'departamento_id', 'name': 'departamento'}, inplace=True)
depto_copy.drop(columns=['tasa_con_nbi'], axis=1, inplace=True)
depto_copy.replace('nc', np.nan, inplace=True)
depto_copy.replace(-1, np.nan, inplace=True)
depto_copy.replace('sin información', np.nan, inplace=True)
loc_df = pd.merge(depto_copy, provincia_df.copy(), left_on='provincia_id', right_on='id')
loc_df.drop(columns=['id'], inplace=True)
loc_df.rename(columns={'name': 'provincia'}, inplace=True)

df = pd.merge(df, loc_df, how='left' ,on=['departamento', 'provincia'])
df.drop(columns=['departamento', 'provincia'], inplace=True)
```
